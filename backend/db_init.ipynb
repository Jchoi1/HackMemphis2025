```python
import sqlite3
```

```python
db = sqlite3.connect('211Memphis.db')
cursor = db.cursor()
```

```python
cursor.execute('''CREATE TABLE users (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    username TEXT UNIQUE NOT NULL,
    password TEXT NOT NULL,
    name TEXT,
    address TEXT,
    org_name TEXT,
    org_address TEXT,
    org_phone TEXT,
    org_email TEXT,
    org_service_description TEXT,
    org_eligibility TEXT,
    org_intake_procedure TEXT,
    org_hours TEXT,
    role TEXT NOT NULL CHECK(role IN ('user', 'organization', 'admin'))
);
''')
```

```python
# test organization user accounts
db.row_factory = sqlite3.Row
cursor = db.cursor()

cursor.execute("SELECT * FROM users WHERE role = 'organization'")
for row in cursor.fetchall():
    print(dict(row))

```
